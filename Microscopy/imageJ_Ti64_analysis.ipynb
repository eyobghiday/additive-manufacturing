```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
# %matplotlib notebook
```

```python
data= pd.read_excel("density.xlsx")
data
```

```python
from matplotlib import rcParams
rcParams['figure.figsize'] = 10,6
plt.style.use('classic')
plt.xlabel('Scan Speed(mm/s)', size = 15)
plt.ylabel('Power(W)', size = 15)
# plt.text(400,150,'(400, 150)')
plt.scatter(data.Scanspeed,data.Power, c = data.Porosity, cmap ='Spectral' , s=175, edgecolor ='k')
cbar=plt.colorbar(orientation = 'vertical');
cbar.set_label("Porosity%",size = 15)
# plt.savefig("graph.svg")
# plt.savefig('filename.pdf')

for x,y in zip(data.Scanspeed,data.Power):

    label = f"(2)"

    plt.annotate(label, # this is the text
                 (x,y), # these are the coordinates to position the label
                 textcoords="offset points", # how to position the text
                 xytext=(0,10), # distance from text to points (x,y)
                 ha='center') # horizontal alignment can be left, right or center
```
